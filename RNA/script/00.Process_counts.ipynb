```python
import pandas as pd
```

```python
counts_full_df = pd.read_csv("../../data/counts.matrix.full.txt", sep='\t', index_col=0)
counts_pooling_df  = pd.read_csv("../../data/counts.pooling.txt", sep='\t', index_col=0)
meta_df   = pd.read_csv("../../data/metadata.310.txt",   sep='\t')
```

```python
cond1 = meta_df['library'] == 'full'
cond2 = meta_df['cell'].str.startswith('WYZX_E4')
target_cells = meta_df.loc[cond1 | cond2, 'cell'].tolist()
print(f"Full：{len(target_cells)}")
```

```python
# smartseq dataset:version 1 with WYZX_E4
target_cells_valid = [c for c in target_cells if c in counts_full_df.columns]
result_df = counts_full_df[target_cells_valid]
print(f"matrix：{result_df.shape}  （genes x cells) ")
result_df.to_csv('../../data/counts.matrix.full.QC.WYZX_E4.104.txt', sep='\t')
```

```python
# smartseq dataset:version 2 without WYZX_E4
target_cells_no_e4 = meta_df.loc[cond1 & ~cond2, 'cell'].tolist()
target_cells_no_e4_valid = [c for c in target_cells_no_e4 if c in counts_full_df.columns]
result_df_no_e4 = counts_full_df[target_cells_no_e4_valid]
print(f"matrix：{result_df_no_e4.shape}  （genes x cells) ")
result_df_no_e4.to_csv('../../data/counts.matrix.full.QC.no_WYZX_E4.81.txt', sep='\t')
```

```python
cond3 = meta_df['library'] == 'pooling'
cond4 = meta_df['cell'].str.startswith('WYZX_E4')
target_cells_pooling = meta_df.loc[cond3, 'cell'].tolist()
print(f"Pooling：{len(target_cells_pooling)}")
```

```python
# strt-seq dataset:version with WYZX_E4
target_cells_pooling_valid = [c for c in target_cells_pooling if c in counts_pooling_df.columns]
result_df_pooling_e4 = counts_pooling_df[target_cells_pooling_valid]
print(f"matrix：{result_df_pooling_e4.shape}  （genes x cells）")

n = result_df_pooling_e4.shape[1]
result_df_pooling_e4.to_csv(f'../../data/counts.matrix.pooling.QC.with_WYZX_E4.{n}.txt', sep='\t')
```

```python
target_cells_pooling_no_e4 = meta_df.loc[cond3 & ~cond4, 'cell'].tolist()
print(f"Pooling(no WYZX_E4), cells:{len(target_cells_pooling_no_e4)}")
```

```python
target_cells_pooling_no_e4_valid = [c for c in target_cells_pooling_no_e4 if c in counts_pooling_df.columns]
result_df_pooling_no_e4 = counts_pooling_df[target_cells_pooling_no_e4_valid]
print(f"matrix：{result_df_pooling_no_e4.shape}  （genes x cells）\n")

n2 = result_df_pooling_no_e4.shape[1]
result_df_pooling_no_e4.to_csv(f'../../data/counts.matrix.pooling.QC.no_WYZX_E4.{n2}.txt', sep='\t')
```
